```python
x = trainset[0][0]
y = trainset[0][1]
```

```python
class cnn1_layer(nn.Module):
    def __init__(self):
        super(cnn1_layer, self).__init__()
        self.conv = nn.Conv2d(in_channels = 1, out_channels = 8, kernel_size=(3,3), padding=(1,1))

    def forward(self, x):
        x = self.conv(x)
        return x
```

```python
model = cnn1_layer()
loss = nn.MSELoss()
optimizer = torch.optim.RMSprop(model.parameters(), lr = 0.001, momentum = 0.9)


for _ in range(50):
    model.train()
    pred = model(x)
    
    # Ensure y is the correct shape and type
    if isinstance(y, int) or len(y.shape) == 0:
        y = torch.tensor(y, dtype=pred.dtype).expand_as(pred)
    else:
        y = y.type_as(pred)
    
    loss_fn = loss(pred, y)
    optimizer.zero_grad()  # Fixed from optimizer.grad_zero()
    loss_fn.backward()
    optimizer.step()

    model.eval()
    with torch.no_grad():  # Avoid computing gradients during evaluation
        test_pred = model(x)
        test_loss = loss(test_pred, y)  # Use a different variable for clarity
```
